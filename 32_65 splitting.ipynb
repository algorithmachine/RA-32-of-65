```python
k = 3
blueset = {x for x in range(2**(3*k+1)) if bin(x).count('1') >= 2*k+1}
redset = {x for x in range(2**(3*k+1)) if 0 < bin(x).count('1') <= 2*k}
```

```python
def sumset(X,Y): return {x^y for x in X for y in Y}
```

```python
def split(X):
    A = {x for x in X if random.random() > 0.5}
    B = X - A
    return A,B
```
